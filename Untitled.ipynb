```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
import warnings
from scipy.stats.stats import pearsonr
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
train=pd.read_csv('E:/崔庆才-机器学习/第一周/house-prices-advanced-regression-techniques/train.csv')
test=pd.read_csv('E:/崔庆才-机器学习/第一周/house-prices-advanced-regression-techniques/test.csv')
```

```python
train.head()
```

```python
all_data=pd.concat((train.loc[:,'MSSubClass':'SaleCondition'],test.loc[:,'MSSubClass':'SaleCondition']))
print(all_data.head())
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
matplotlib.rcParams['figure.dpi']=300
prices=pd.DataFrame({'price':train['SalePrice'],'log(price+1)':np.log1p(train['SalePrice'])})
prices.hist()
```

```python
from scipy.stats import skew
train['SalePrice']=np.log1p(train['SalePrice'])
numeric_feats=all_data.dtypes[all_data.dtypes!='object'].index
#print(numeric_feats)
skewed_feats=train[numeric_feats].apply(lambda x:skew(x.dropna()))
skewed_feats=skewed_feats[skewed_feats>0.75]
skewed_feats=skewed_feats.index
all_data[skewed_feats]=np.log1p(all_data[skewed_feats])
print(all_data[skewed_feats].head())
```

```python
all_data=pd.get_dummies(all_data)
all_data=all_data.fillna(all_data.mean())
all_data.head()
```

```python
x_train=all_data[:train.shape[0]]
x_test=all_data[train.shape[0]:]
y=train.SalePrice
```

```python
train.shape[1]
```

```python
from sklearn.linear_model import Ridge,RidgeCV,ElasticNet,LassoCV,LassoLarsCV
from sklearn.model_selection import cross_val_score
def rmse_cv(model):
    rmse=np.sqrt(-cross_val_score(model,x_train,y,scoring='neg_mean_squared_error',cv=5))
    return rmse
```

```python
model_ridge=Ridge()
```

```python
alphas=[0.05,0.1,0.3,1,3,5,10,15,30,50,75]
cv_ridge=[rmse_cv(Ridge(alpha=alpha)).mean() for alpha in alphas]
```

```python
cv_ridge=pd.Series(cv_ridge,index=alphas)
cv_ridge.plot(title='Validation-Just Do It')
plt.xlabel('alpha')
plt.ylabel('rmse')
```

```python
cv_ridge.min()
```

```python
model_lasso=LassoCV(alphas=[1,0.1,0.001,0.0005]).fit(x_train,y)
```

```python
rmse_cv(model_lasso).mean()
```

```python
coef=pd.Series(model_lasso.coef_,index=x_train.columns)
print(coef)
```

```python
print('Lasso picked '+str(sum(coef!=0))+' varialbes and eliminated the other '+str(sum(coef==0))+' variables')
```

```python
imp_coef=pd.concat([coef.sort_values().head(10),
                   coef.sort_values().tail(10)])
```

```python
matplotlib.rcParams['figure.figsize']=(8,10)
imp_coef.plot(kind='barh')
plt.title('Coefficients in the Lasso Model')
```

```python
matplotlib.rcParams['figure.figsize']=(6,8)
preds=pd.DataFrame({'preds':model_lasso.predict(x_train),"true":y})
preds['residuals']=preds['true']-preds['preds']
preds.plot(x='preds',y='residuals',kind='scatter')
```

```python
import matplotlib.pyplot as plt
x_train=[100,80,120,75,60,43,140,132,63,55,74,44,88]
y_train=[120,92,143,87,60,50,167,147,80,60,90,57,99]
alpha=0.00001
cnt=0
m=len(x_train)
theta0=0
theta1=0
error0=0
error1=0
epsilon=0.000001
def h(x):
    return theta1*x+theta0
while True:
    cnt=cnt+1
    diff=[0,0]
    for i in range(m):
        diff[0]+=h(x_train[i])-y_train[i]
        diff[1]+=(h(x_train[i])-y_train[i])*x_train[i]
    theta0=theta0-alpha/m*diff[0]
    theta1=theta1-alpha/m*diff[1]
    error1=0
    for i in range(len(x_train)):
        error1+=(y_train[i]-(theta0+theta1*x_train[i]))*2*2/2
    if abs(error1-error0)<epsilon:
        break
    else:
        error0=error1
plt.plot(x_train,[h(x) for x in x_train])
plt.plot(x_train,y_train,'bo')
print(theta1,theta0)
plt.show()
```
